## Задание 6

Модифицируйте функцию, осуществляющую метод градиентного спуска с семинара: добавьте опцию останавливать метод, если абсолютное значение производной функции в точке < некоторого $\delta$ (delta добавьте в аргументы функции).

```python
x = sp.Symbol('x', real=True)
```

```python
def gradient_descent(funct, x_start, learning_rate, epsilon, num_iterations, delta):
    x_curr = x_start
    df_x = sp.diff(funct)

    trace = []
    trace.append(x_curr)

    for i in range(num_iterations):
        x_new = x_curr - learning_rate * df_x.subs(x, x_curr)
        trace.append(x_new)

        if (abs(x_new - x_curr) < epsilon) or (abs(df_x.subs(x, x_curr)) < delta):
            return x_curr, trace

        x_curr = x_new

    return x_curr, trace
```

## Задание 8

Реализуйте метод Ньютона (метод второго порядка) для поиска минимума функции.

Про метод ещё можно почитать [здесь](https://math.semestr.ru/optim/minimum.php).

```python
def newtons_method(funct, a, b, epsilon, num_iterations):

    df_x = sp.diff(funct)
    df2_x = sp.diff(df_x)
    df3_x = sp.diff(df2_x)

    if df_x.subs(x, a) * df3_x.subs(x, a) > 0:
        x_start = a
    else:
        x_start = b

    x_curr = x_start
    trace = []
    trace.append(x_curr)

    for i in range(num_iterations):
        x_new = x_curr - df_x.subs(x, x_curr) / df2_x.subs(x, x_curr)
        trace.append(x_new)

        if abs(x_new - x_curr) < epsilon:
            return x_curr, trace, x_start

        x_curr = x_new

    return x_curr, trace, x_start
```

## Задание 11*

Реализуйте оптимизацию метода градиентного спуска - Momentum.

```python
def momentum(funct, x_start, learning_rate, epsilon, num_iterations, delta, betta):
    x_curr = x_start
    df_x = sp.diff(funct)

    trace = []
    trace.append(x_curr)

    x_prev = x_curr

    for i in range(num_iterations):
        x_new = x_curr - learning_rate * df_x.subs(x, x_curr) + betta * (x_curr - x_prev)
        x_prev = x_curr
        trace.append(x_new)

        if (abs(x_new - x_curr) < epsilon) or (abs(df_x.subs(x, x_curr)) < delta):
            return x_curr, trace

        x_curr = x_new

    return x_curr, trace
```
